```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
```

```python
# Sample Data
df = pd.DataFrame ({
    'Income' : [40, 45, 47, 50, 18, 20, 22, 80, 82, 85],
    'Score' : [30, 32, 34, 36, 60, 62, 64, 20, 18, 15]
})
```

```python
# Dendrogram using Euclidean Distance
plt.figure(figsize=(8,5))
dendrogram = sch.dendrogram(
    sch.linkage(df, method = 'complete', metric = 'euclidean')
)
plt.title("Dendrogram (Euclidean Distance)")
plt.xlabel("Customers")
plt.ylabel("Distance")
plt.show()
```

```python
# Hierarchical Clustering Using Euclidean Distance
hc = AgglomerativeClustering(
    n_clusters = 3,
    metric = 'euclidean',  # Euclidean distance
    linkage = 'single'     # single, complete, average
)
df['Cluster'] = hc.fit_predict(df)
print(df)
```

----

Divisive Clustering using Bisecting K-Means

```python
from sklearn.cluster import BisectingKMeans

bisect = BisectingKMeans(n_clusters = 3, random_state = 42)
labels = bisect.fit_predict(df[['Income', 'Score']])
```

```python
plt.scatter(df['Income'], df['Score'], c = labels, s = 50)
plt.show()
```
